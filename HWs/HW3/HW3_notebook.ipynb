let's define a function to make accessing countries easier


```python
all_countries = wbdata.get_countries()

country_dict = {country["id"]: country["name"] for country in all_countries}

def get_country_names_list(codes):
    return [country_dict.get(code, "Unknown Country") for code in codes]
```
