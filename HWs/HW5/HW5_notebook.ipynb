```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

In this assignment we'll use the diabetes dataset that's been proccessed and feature engineered in the last assignment.

```python
import pandas as pd 
df = pd.read_csv('diabetes.csv')
```

### Regression

we'll now preapre our dataset for a simple regression and test out the metrics that we want on our model.

```python
diabetes = df.pop('diabetes')
```

```python
from sklearn.preprocessing import StandardScaler
normalized_df = StandardScaler().fit_transform(df)
```

```python
from sklearn import linear_model
model = linear_model.Ridge(alpha = 300)

model.fit(normalized_df, diabetes)
```

all is ready ! <br>
lets test the metrics.

### MSE :

```python
from sklearn.metrics import mean_squared_error
diabetes_prediction = model.predict(normalized_df)
MSE = mean_squared_error(diabetes, diabetes_prediction)
print('MSE is %.4f' %MSE)
```

### MAE : 

```python
from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(diabetes , diabetes_prediction)
print('MAE is %f' %MAE)
```

### MAPE :

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE = mean_absolute_percentage_error(diabetes, diabetes_prediction)
print('MAPE is %.4f' %MAPE)
```

not such a good MAPE ! <br>
probably because of the binary nature of our test data.

### R2 Score :

```python
from sklearn.metrics import r2_score
R2_score = r2_score(diabetes, diabetes_prediction)
print('R2 Score is %.4f' %R2_score)
```

### Binary Classification

let's now do the same task with a simple decision tree method.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(normalized_df, diabetes,test_size=.5)
```

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier(max_depth=10)
classifier.fit(X_train,y_train)
```

we're set to test !

```python
from sklearn.metrics import accuracy_score
predicted_diabetes = classifier.predict(X_test)
print('Calssification accruacy is %.2f' %accuracy_score(predicted_diabetes,y_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(predicted_diabetes,y_test)
```

### Recall & Precision :

```python
from sklearn.metrics import precision_score, recall_score
print("Recall Score is : %f" %recall_score(y_pred=predicted_diabetes , y_true=y_test))
print("Precision Score is : %f" %precision_score(y_pred=predicted_diabetes , y_true=y_test))
```

### F1 Score :

```python
from sklearn.metrics import f1_score
f1s = f1_score(y_test, predicted_diabetes)
print('Macro-averaged F1-score is %.2f' %f1s)
```

### Multi Class Classifcation


let's do a multi-class classification for a column in our data called heart_problem_rate.

```python
hpr = df.heart_problem_rate
df.pop('heart_problem_rate')
df.pop('heart_disease')
df.pop('hypertension')
normalized_df = StandardScaler().fit_transform(df)
```
